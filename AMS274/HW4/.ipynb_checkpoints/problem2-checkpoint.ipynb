```python

with pm.Model() as modelo_s:
    alfa = pm.Normal('alfa', mu=0, sd=10, shape=3)
    beta = pm.Normal('beta', mu=0, sd=10, shape=(2,3))

    mu = alfa + T.dot(x,beta)
    p = pm.Deterministic('p', T.nnet.softmax(mu))

    yl = pm.Categorical('yl', p=p, observed=y)
    step = pm.HamiltonianMC()
    trace_s = pm.sample(5000, step=step, chains=1)
```

```python
pm.traceplot(trace_s,varnames=['alfa', 'beta'])
```
